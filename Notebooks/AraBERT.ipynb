```python
import pandas as pd
import time
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python

df = pd.read_csv("Saudi.csv").sample(n=1600, random_state=42).reset_index(drop=True)
```

```python
arabert_model = pipeline("text-classification",
    model="akhooli/xlm-r-large-arabic-toxic",
    tokenizer="akhooli/xlm-r-large-arabic-toxic")

def arabert_classify(text):
    try:
        result = arabert_model(text)
        return result[0]["label"].lower()
    except Exception as e:
        return f"error: {e}"


print("Annotating with AraBERT...")
df["arabert"] = df["tweet"].apply(arabert_classify)


```

```python
label_map = {
    "label_0": "not",
    "label_1": "offensive"
}
df["arabert_mapped"] = df["arabert"].map(label_map)
```

```python
df.to_csv("ArabBERT.csv", index=False)
```
